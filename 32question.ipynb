```python
#1  Reverse an array
# Method 1: Slicing (Create a New Array)
# The most common and "Pythonic" way is to use list slicing with a step of -1.
# Logic: The syntax [::-1] tells Python to start from the end and move backwards to the beginning with a step of -1.
# This creates a copy of the array, leaving the original unchanged
arr=[1,2,3,4,5]
print(arr)
rever_arr=arr[::-1]
print(rever_arr)
```

```python
#Method 2: In-place Reversal (Modify the Original)

#If you want to save memory and reverse the array directly, use the built-in reverse() method.
#Logic: This method modifies the original list in-place and does not return anything.
#It is highly efficient for large datasets because it doesn't require extra space to store a new copy.
arr = [1, 2, 3, 4, 5]
arr.reverse()
print(arr)
```

```python
# method 3: manual  Two-Pointer Technique
# both pointers will stop when they meet each other or cross each other .
def reverse_array(arr):
    start = 0
    end = len(arr) - 1
    while start < end:
        arr[start], arr[end] = arr[end], arr[start]
        start += 1
        end -= 1
    return arr

arr = [1, 2, 3, 4, 5]
print(reverse_array(arr))
```

```python
#3. Find the Sum of Elements

def sum_of_elements(arr):
    total = 0
    for num in arr:
        total += num
    return total

arr = [1, 2, 3, 4, 5]
print(sum_of_elements(arr))

```

```python
# 4. Find the Second Largest Element

def second_largest(arr):
    first = second = float('-inf')

    for num in arr:
        if num > first:
            second = first
            first = num
        elif num > second and num != first:
            second = num

    return second

numbers = [10, 20, 4, 45, 99]
print(second_largest(numbers))
```

```python
#5. Count Frequency of Elements

def count_frequency(arr):
    freq = {}

    for num in arr:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1

    return freq

numbers = [1, 2, 2, 3, 3, 3, 4]
print(count_frequency(numbers))

```

```python
#6. Check if Array is Sorted

def is_sorted(arr):
    for i in range(len(arr) - 1):
        if arr[i] > arr[i + 1]:
            return False
    return True

numbers = [1, 2, 3, 6, 5]
print(is_sorted(numbers))


```

```python
#7. Rotate Array by k Positions: Rotate the array to the right by k positions.

def rotate_array(arr, k):
    n = len(arr)
    k = k % n   # handle k > n
    return arr[-k:] + arr[:-k]

numbers = [1, 2, 3, 4, 5]
print(rotate_array(numbers, 2))
```

```python
#8. Find Pair with Given Sum: Find a pair of elements that adds up to a target sum.

def find_pair(arr, target):
    seen = set()

    for num in arr:
        complement = target - num
        if complement in seen:
            return (complement, num)
        seen.add(num)

    return None

numbers = [2, 7, 11, 15]
target = 9
print(find_pair(numbers, target))
```

```python
#9. Remove Duplicates from Array: Remove duplicates from the array while maintaining order.

def remove_duplicates(arr):
    seen = set()
    result = []

    for num in arr:
        if num not in seen:
            seen.add(num)
            result.append(num)

    return result

numbers = [1, 2, 2, 3, 4, 3, 5]
print(remove_duplicates(numbers))
```

```python
#10.Merge Two Sorted Arrays

def merge_sorted_arrays(arr1, arr2):
    i = j = 0
    merged = []

    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j]:
            merged.append(arr1[i])
            i += 1
        else:
            merged.append(arr2[j])
            j += 1

    # Add remaining elements
    merged.extend(arr1[i:])
    merged.extend(arr2[j:])

    return merged

arr1 = [1, 3, 5]
arr2 = [2, 4, 6]
print(merge_sorted_arrays(arr1, arr2))
```

```python
#12 Find the Missing Number: Find the missing number in an array of size n containing numbers from 1 to n.

def find_missing(arr, n):
    xor_all = 0
    xor_arr = 0

    for i in range(1, n + 1):
        xor_all ^= i

    for num in arr:
        xor_arr ^= num

    return xor_all ^ xor_arr


# Example
arr = [1, 2, 4, 5]
n = 5
print(find_missing(arr, n))  # Output: 3
```

```python
#13 Find Duplicates in an Array

def find_duplicates(arr):
    seen = set()
    duplicates = set()

    for num in arr:
        if num in seen:
            duplicates.add(num)
        else:
            seen.add(num)

    return list(duplicates)

numbers = [1, 2, 3, 2, 4, 5, 1]
print(find_duplicates(numbers))
```

```python
#17 Find the Leader Elements: An element is a leader if it is greater than all elements to its right.

def find_leaders(arr):
    leaders = []
    max_right = float('-inf')


    for num in reversed(arr):
        if num > max_right:
            leaders.append(num)
            max_right = num

    return leaders[::-1]

arr = [16, 17, 4, 3, 5, 2]
print(find_leaders(arr))
```

```python
#18 Move Zeroes to End: Move all zeroes in an array to the end while maintaining the order of non-zero elements.

def move_zeroes(arr):
    j = 0  # position for next non-zero

    for i in range(len(arr)):
        if arr[i] != 0:
            arr[j] = arr[i]
            j += 1

    # Fill remaining positions with zero
    for i in range(j, len(arr)):
        arr[i] = 0

    return arr

numbers = [0, 1, 0, 3, 12]
print(move_zeroes(numbers))
```

```python
#19 Find Subarray with Given Sum.

def subarray_sum(arr, target):
    left = 0
    current_sum = 0

    for right in range(len(arr)):
        current_sum += arr[right]

        # Shrink window if sum exceeds target
        while current_sum > target:
            current_sum -= arr[left]
            left += 1

        if current_sum == target:
            return (left, right)  # indices of subarray

    return None

arr = [1, 2, 3, 7, 5]
target = 12

print(subarray_sum(arr, target))


```

```python
#23 Maximum Sum Subarray (Kadane's Algorithm)

def max_subarray_sum(arr):
    max_sum = arr[0]
    current_sum = arr[0]

    for i in range(1, len(arr)):
        current_sum = max(arr[i], current_sum + arr[i])
        max_sum = max(max_sum, current_sum)

    return max_sum

arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
print(max_subarray_sum(arr))
```

```python
#24 Rearrange Array Alternately: Rearrange an array such that elements alternate between the largest and smallest.

def rearrange(arr):
    arr.sort()
    result = []

    left, right = 0, len(arr) - 1

    while left <= right:
        if left != right:
            result.append(arr[right])  # largest
            result.append(arr[left])   # smallest
        else:
            result.append(arr[left])   # middle element

        left += 1
        right -= 1

    return result

arr = [1, 2, 3, 4, 5, 6]
print(rearrange(arr))
```

```python
#25 Find Majority Element: Find the element that appears more than n/2 times.

def majority_element(arr):
    candidate = None
    count = 0

    # Step 1: Find candidate
    for num in arr:
        if count == 0:
            candidate = num
        count += (1 if num == candidate else -1)

    # Step 2: Verify candidate (optional but safe)
    if arr.count(candidate) > len(arr) // 2:
        return candidate

    return None

arr = [2, 2, 1, 1, 2, 2, 2]
print(majority_element(arr))
```

```python
#26 Find Peak Element: A peak element is greater than its neighbors. Find one such element.

def find_peak(arr):
    n = len(arr)

    for i in range(n):
        if (i == 0 or arr[i] >= arr[i-1]) and \
           (i == n-1 or arr[i] >= arr[i+1]):
            return i   # return index of peak

    return -1

arr = [1, 3, 20, 4, 1, 0]
print(find_peak(arr))
```

```python
#27 Find the First Missing Positive: Find the smallest positive integer missing in the array.

def first_missing_positive(nums):
    n = len(nums)


    for i in range(n):
        while 1 <= nums[i] <= n and nums[nums[i] - 1] != nums[i]:
            correct_index = nums[i] - 1
            nums[i], nums[correct_index] = nums[correct_index], nums[i]


    for i in range(n):
        if nums[i] != i + 1:
            return i + 1


    return n + 1



arr = [3, 4, -1, 1]
print(first_missing_positive(arr))
```

```python
#28 Sort an Array of 0s, 1s, and 2s

def sort_012(arr):
    low = 0
    mid = 0
    high = len(arr) - 1

    while mid <= high:
        if arr[mid] == 0:
            arr[low], arr[mid] = arr[mid], arr[low]
            low += 1
            mid += 1

        elif arr[mid] == 1:
            mid += 1

        else:
            arr[mid], arr[high] = arr[high], arr[mid]
            high -= 1

    return arr



arr = [2, 0, 2, 1, 1, 0]
print(sort_012(arr))
```

```python
#29 Find the Longest Consecutive Sequence

def longest_consecutive(nums):
    num_set = set(nums)
    longest = 0

    for num in num_set:

        if num - 1 not in num_set:
            current = num
            count = 1

            while current + 1 in num_set:
                current += 1
                count += 1

            longest = max(longest, count)

    return longest



arr = [100, 4, 200, 1, 3, 2]
print(longest_consecutive(arr))
```

```python
#30. Product of Array Except Self
#Given an array, return a new array where each element is the product of all elements except itself.
#Do not use division. Input: [1,2,3,4] Output: [24,12,8,6]'''

def product_except_self(nums):
    n = len(nums)
    result = [1] * n


    prefix = 1
    for i in range(n):
        result[i] = prefix
        prefix *= nums[i]


    suffix = 1
    for i in range(n - 1, -1, -1):
        result[i] *= suffix
        suffix *= nums[i]

    return result



arr = [1, 2, 3, 4]
print(product_except_self(arr))

```

```python
# 31 Find Equilibrium Index: Find an index such that sum of elements on left = sum on right.
def equilibrium_index(arr):
    n = len(arr)
    prefix = [0] * n

    prefix[0] = arr[0]
    for i in range(1, n):
        prefix[i] = prefix[i-1] + arr[i]

    for i in range(n):
        left = prefix[i-1] if i > 0 else 0
        right = prefix[n-1] - prefix[i]

        if left == right:
            return i

    return -1


```

```python
# 32 Find Maximum Product Pair: Find two elements whose product is maximum.

def max_product_pair(arr):
    n = len(arr)
    if n < 2:
        return None


    max1 = max2 = float('-inf')
    min1 = min2 = float('inf')


    for num in arr:

        if num > max1:
            max2 = max1
            max1 = num
        elif num > max2:
            max2 = num


        if num < min1:
            min2 = min1
            min1 = num
        elif num < min2:
            min2 = num


    if max1 * max2 > min1 * min2:
        return (max1, max2)
    else:
        return (min1, min2)



arr = [-10, -3, 5, 6, -2]
print(max_product_pair(arr))

```
